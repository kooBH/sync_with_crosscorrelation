```python
import sync
import numpy as np
```

```python
path_base = 'D:/VoiceFilter/chime4_MLDR/'
path_clean = 'clean/dt05_bth_0/'
path_mldr = 'mldr/dt05_bus_simu/'
```

```python

import wave

mixed = wave.open(path_base+path_clean+'F01_22GC010A_BTH.CH0.wav','r')
clean = wave.open(path_base+path_mldr+'F01_22GC010A_BUS.wav','r')

#len_1 = mixed.getnframes()
#len_2 = clean.getnframes()

data_1 = mixed.readframes(-1)
data_2 = clean.readframes(-1)

data_1 = np.fromstring(data_1, "Int16")
data_2 = np.fromstring(data_2, "Int16")
```

```python

from matplotlib import pyplot as plt

plt.plot(data_1)
plt.show()

plt.plot(data_2)
plt.show()

plt.plot(data_1-data_2)
plt.show()
```
